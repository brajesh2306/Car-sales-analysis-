```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('/content/CarPrice_Assignment.csv')
```

```python
def car_name(text):
  return text.split(' ')[0]
```

```python
data['CarName'] = data['CarName'].apply(car_name)
```

```python
data
```

```python
data.info()
```

```python
data.columns
```

```python
data['CarName'].unique()
```

```python
def spell_correct(text,correct_text):
  data['CarName'].replace(text,correct_text,inplace=True)
```

```python
spell_correct('maxda','mazda')
spell_correct('Nissan','nissan')
spell_correct('porcshce','porsche')
spell_correct('toyouta','toyota')
spell_correct('vokswagen','volkswagen')
spell_correct('vw','volkswagen')
```
